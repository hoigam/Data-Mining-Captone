```python
import os
from sklearn.feature_extraction.text import TfidfVectorizer
```

## Solution1: TF-IDF + Logistic Regeression

```python
base_path = '/kaggle/working/Hygiene'

reviews_path = os.path.join(base_path, 'hygiene.dat')
labels_path = os.path.join(base_path, 'hygiene.dat.labels')
```

```python
N = 546

with open(reviews_path, 'r') as f:
    data_train = [next(f) for x in range(N)]
    data_test = [x for x in f]
    
with open(labels_path, 'r') as f:
    y_train = [next(f) for x in range(N)]  
```

```python
y_train = [x.replace('\n', '') for x in y_train]
```

```python
tf_idf_vec = TfidfVectorizer(stop_words='english')
```

```python
data_test_vec = tf_idf_vec.transform(data_test)
```

```python
data_test_vec.shape
```

```python
data_train_vec = tf_idf_vec.fit_transform(data_train)
data_train_vec.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=1875)
```

```python
clf.fit(data_train_vec, y_train)
```

```python
clf.predict(data_train_vec)
```

```python
clf.predict_proba(data_train_vec)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_train, clf.predict_proba(data_train_vec)[:, 1])
```

```python
test_pred = clf.predict(data_test_vec)
```
